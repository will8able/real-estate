```python
from bs4 import BeautifulSoup
import requests
import re
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
```

```python
def get_HTML_document(url):
    # request for HTML document of given url
    response = requests.get(url)
    # response will be provided in JSON format
    return response.text

def get_strings_regex(url_to_scrape, attribute, pattern = ''):
    # create document
    html_document = get_HTML_document(url_to_scrape)

    # create soap object
    soup = BeautifulSoup(html_document, 'html.parser')
    
    attr_dict = dict()
    attr_dict[attribute] = re.compile(pattern)
    
    str_list = []

    # find all the anchor tags with attribute
    # based on regex pattern
    for link in soup.find_all('a', attrs=attr_dict):
        # display the actual urls
        str_list.append(link.get(attribute))
    return str_list
```

```python
a_streets_url = "http://ldsnet.fairfaxcounty.gov/landusesrch/LUStreets.aspx?startswith=A"

a_street_url2_list = get_strings_regex(a_streets_url, 'href', '^Default')
a_street_url2_list
```
